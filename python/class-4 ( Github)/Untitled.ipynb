# Git:
Git is a distributed version control system (VCS).
It is a software tool installed and run locally on a developer's machine.
Git's primary function is to track changes in source code and other files over time, allowing developers to revert to previous versions, manage different branches of development, and merge changes.
It operates independently, meaning you can use Git for version control without needing an online platform.
# GitHub:
GitHub is a cloud-based hosting service for Git repositories.
It is an online platform that leverages Git to provide collaborative features and a centralized location for storing and managing code.
GitHub extends Git's core functionality with features such as pull requests, issue tracking, project management tools, and a social networking aspect for developers.
While GitHub uses Git as its underlying technology, it is not a version control system itself.
In essence:
Git is the engine: that handles version control locally.
GitHub is the garage: (or cloud-based platform) that hosts those Git-managed projects and facilitates collaboration among developers.
You can use Git without GitHub, but you cannot use GitHub without Git. GitHub serves as a popular and feature-rich online environment for teams to collaborate on projects managed with Git. 
Git Init vs Git Clone
# 1 Git Init

git init is used to create a brand-new Git repository in your local machine.

It turns an existing folder (with your project files) into a Git repository.

Useful when starting a new project from scratch.
# How it works (example):
- Go to the folder where your project exists
cd /d/cde-3/python  

# Initialize Git in this folder
- git init  

# Add all files for tracking
- git add .  

# Commit the changes
- git commit -m "Initial commit"

# Link to your GitHub repo
- git remote add origin https://github.com/qayasmehtab/repo-name.git  

# Push files to GitHub
- git branch -M main 
- git push -u origin main

# 2 Git Clone

- git clone is used to copy an existing GitHub repository to your local machine.

- It downloads the full project (with history, branches, and commits).
